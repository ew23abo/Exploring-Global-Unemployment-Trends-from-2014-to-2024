```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Uploading and Reading the Dataset
df = pd.read_csv ('global_unemployment_data.csv')

print(df)
```

```python
#Defining the dataframe and columns
df.info()
```

```python
# Total null values in the dataframe
df.isnull().sum()
```

```python
# Count the number of missing values in each column
missing_values = df.isnull().sum()

# Remove rows with any missing values
cleaned_df = df.dropna()

# Print the cleaned DataFrame
print(cleaned_df)
```

```python
# Filter out rows where age_group is not "Under 15"
df_filtered = df[df['age_group'] != 'Under 15']
```

```python
# Now df_filtered contains the dataset with "Under 15" age group removed
df_filtered
```

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Calculate average unemployment rate by year for all, male, and female


avg_unemployment_male = df[df['sex'] == 'Male'][['2014', '2015', '2016', '2017', '2018', '2019', 
                                                 '2020', '2021', '2022', '2023', '2024']].mean()
avg_unemployment_female = df[df['sex'] == 'Female'][['2014', '2015', '2016', '2017', '2018', '2019',
                                                     '2020', '2021', '2022', '2023', '2024']].mean()


# Plot the overall average unemployment rate
plt.figure(figsize=(10, 5))

# Plot the average unemployment rate for males
avg_unemployment_male.plot(kind='line', marker='o', color='blue', label='Male')

# Plot the average unemployment rate for females
avg_unemployment_female.plot(kind='line', marker='*', color='red', label='Female')

plt.title('Average Unemployment Rate by Gender (2014 - 2024)', fontweight = 'bold')
plt.ylabel('Average Unemployment Rate (%)')
plt.legend()  # Show legend

plt.grid(True)

# Add figure explanation (caption) below the graph
plt.figtext(0.5, 0.01, "Figure 1: This line plot illustrates the average unemployment rate by gender from 2014 to 2024.", ha='center', fontsize=10, color = 'black', fontweight = 'bold')


plt.show()
```

```python
# Compare unemployment rates across gender and age groups
plt.figure(figsize=(10,5.5))
sns.barplot(x='sex', y='2020', hue = 'age_group',  data = df_filtered)

plt.title('Unemployment Rate by Gender and Age Group in 2020', fontweight= 'bold')
plt.xlabel('Gender')
plt.ylabel('Unemployment Rate (%)')

# Add figure explanation (caption) below the graph
plt.figtext(0.5, 0.01, "Figure 2: Comparison of unemployment rates by gender and age group in 2020."
            , ha='center', fontsize=10, color='black', fontweight= 'bold')

plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis


# Reshape the data for male and female unemployment rates
male_data = df[df['sex'] == 'Male'][['2014', '2015', '2016', '2017', '2018', '2019', '2020', 
                                     '2021', '2022', '2023', '2024']]
female_data = df[df['sex'] == 'Female'][['2014', '2015', '2016', '2017', '2018', '2019', 
                                         '2020', '2021', '2022', '2023', '2024']]


# Melt the DataFrame to have 'years' and 'unemployment_rate' columns
male_data_melted = male_data.melt(var_name='year', value_name='unemployment_rate')
female_data_melted = female_data.melt(var_name='year', value_name='unemployment_rate')

# Combine male and female data for box plot
combined_data = pd.concat([male_data_melted.assign(gender='Male'), female_data_melted.assign(gender='Female')])

# Plot the box plot
plt.figure(figsize=(10, 8))
sns.boxplot(x='year', y='unemployment_rate', data=combined_data, palette='muted')
plt.title('Unemployment Rate Distribution (2014 - 2024)', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)

# Add figure explanation (caption) below the graph
plt.figtext(0.5, 0.01, "Figure 3: This box plot illustrates the distribution of unemployment rates by gender from 2014 to 2024."
            , ha='center', fontsize=10, fontweight= 'bold', color='black')

# Calculate skewness and kurtosis
skewness = skew(combined_data['unemployment_rate'])
kurt = kurtosis(combined_data['unemployment_rate'])

print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")

plt.show()

# Calculate skewness and kurtosis
skewness = skew(combined_data['unemployment_rate'])
kurt = kurtosis(combined_data['unemployment_rate'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter the DataFrame to get the top 3 countries with the highest unemployment rates in 2020.
top_3_high_unemployment_countries = df.nlargest(3, '2020')

plt.figure(figsize=(10, 6))
ax = top_3_high_unemployment_countries['2020'].plot(kind='bar', color='lightblue')
plt.title('Top 3 Countries with the Highest Average Unemployment Rates in 2020', fontweight='bold')
plt.xlabel('Country', fontweight='bold')
plt.ylabel('Average Unemployment Rate (%)', fontweight='bold')
plt.xticks(rotation=10, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels

# Add percentage labels on top of each bar
for i, v in enumerate(top_3_high_unemployment_countries['2020']):
    ax.text(i, v + 0.5, f'{v:.2f}%', ha='center', va='bottom', fontsize=10, fontweight='bold', color='black')

# Add figure explanation (caption) below the graph
plt.figtext(0.5, 0.01, "Figure 4: This bar chart illustrates the average unemployment rates for the top 3 countries with the highest unemployment rates in 2020.",
            ha='center', fontsize=10, fontweight='bold', color='black')

plt.show()
```

```python
import seaborn as sns
import numpy as np

# Exclude non-numeric columns for correlation analysis
numeric_columns = df.select_dtypes(include='number').columns
correlation_matrix = df[numeric_columns].corr()

# Round the correlation values to 2 decimal places
correlation_matrix = correlation_matrix.round(2)

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, annot=True, cmap='vlag', mask=mask)
plt.title('Pearson Correlation Matrix', fontsize=14, fontweight='bold', ha='center')


# Rotate x-axis and y-axis labels by 45 degrees
plt.xticks(rotation=45)
plt.yticks(rotation=45)

# Add figure explanation (caption) below the graph
plt.figtext(0.5, 0.01, "Figure 5: The heatmap illustrates the Pearson correlation coefficients between unemployment rates and year.", 
            ha='center', fontsize=10, fontstyle = 'italic', color = 'black', fontweight='bold')

plt.show()
```
